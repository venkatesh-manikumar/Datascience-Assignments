4.	What are the ways to combine the dataframe in pandas. Give an example for each

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'Value': ['val0', 'val1', 'val2', 'val3', 'val4', 'val5']})

df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})

print(df1)
print(df2)
```

```python
# approach 1

df1.join(df2, lsuffix='_left', rsuffix='_right')
```

```python
# approach 2

df1.merge(df2, left_on='key', right_on='key')
```

```python
# approach 3

pd.concat([df1, df2])
```

6.	Find out the mean, median and standard deviation of this numpy array -> np.array([1,5,3,100,4,48])

```python
import numpy as np
arr = np.array([1,5,3,100,4,48])
print("MEAN =" , np.mean(arr))
print("MEDIAN =", np.median(arr))
print("STANDARD DEVIATION", np.std(arr))
```

7.	How will you remove duplicate elements from a list?

```python
# loading data in to Set()

list_1 = [1, 2, 1, 4, 6]

print(list(set(list_1)))
```

```python
# using np.unique() 

np.unique(np.array(list_1))
```

```python
# List comprehension

result = [] 
[result.append(x) for x in list_1 if x not in result] 
result
```

```python
# OrderedDict

from collections import OrderedDict 
list(OrderedDict.fromkeys(list_1)) 
```

8.	What is a map() function in Python? Give an example



Python’s map() is a built-in function that allows you to process and transform all the items in an iterable without using an explicit for loop, a technique commonly known as mapping. map() is useful when you need to apply a transformation function to each item in an iterable and transform them into a new iterable.

```python
data = ["map", "example", "demo", "data"]
list(map(str.capitalize, data))
```

10.	What is vstack() in numpy? Give an example

```python
# create two 2d arrays
ar1 = np.array([[1, 2, 3, 4],
                [5, 6, 7, 8]])
ar2 = np.array([[0, 0, 0, 0],
                [1, 1, 1, 1]])

# vstack the arrays
result = np.vstack((ar1, ar2))

# display the concatenated array
result
```
