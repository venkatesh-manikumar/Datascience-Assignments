```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/DataScience/Auto.csv')

data.head()
```

Note: Use automobile dataset for below questions
1. Find the most expensive car company name
2. Count total cars per company
3. Find each company’s highest price car
4. Sort all cars by Price column
5. Group by make, fuel-type and body-style to calculate the average of city-mpg and highway-mpg (miles per gallon)

```python
#1. Find the most expensive car company name
data[data.price == data['price'].max()]
```

```python
#2. Count total cars per company
data.make.value_counts().head()
```

```python
#3. Find each company’s highest price car
data.groupby('make').price.max().head()
```

```python
#4.Sort all cars by Price column
data.sort_values(by=['price'], inplace=True)
data.head()

```

```python
#5. Group by make, fuel-type and body-style to calculate the average of city-mpg and highway-mpg (miles per gallon)

import numpy as np                                                                
data.groupby(by = ['make','fuel_type','body_style']).agg( avg_city_mpg = ('city_mpg',np.mean) , highway_mpg = ('highway_mpg',np.mean) ).reset_index()
                                                                                   
```
